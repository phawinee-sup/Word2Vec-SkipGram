```python
import re

def preprocess_text(text):
    text = re.sub(r"([.,!?])", r" \1 ", text) # split punt (, . ! ?)
    text = re.sub(r"[“”()!?. ]+", r" ", text) # remove non alphabet and puct
    text = re.sub(r'[a-zA-Z]+', lambda x: x.group(0).lower(), text) # Convert English words to lowercase
    text = text.strip()
    return text

cleaned_sentences = [preprocess_text(sentence) for sentence in sentences]
```

```python
from tqdm.notebook import tqdm
import nltk

# Create windows
flatten = lambda outer_list: [item for inner_list in outer_list for item in inner_list]

window_size = 5
windows = flatten([list(nltk.ngrams([MASK_TOKEN] * window_size +
                                    word_tokenize(sentence) + \
                                    [MASK_TOKEN] * window_size, window_size * 2 + 1)) \
          for sentence in tqdm(cleaned_sentences)])
```

```python
data = []
for window in tqdm(windows):
    target_token = window[window_size]
    context = []

    # context are not [MASK] and target_token
    for i, token in enumerate(window):
        if token == MASK_TOKEN or i == window_size or len(token) <= 1:
            continue
        else:
            context.append(token)
    data.append([' '.join(token for token in context), target_token])
```

```python
class Vocabulary(object):
    """Class to process text and extract vocabulary for mapping"""

    def __init__(self, token_to_idx=None, mask_token="<MASK>", add_unk=True, unk_token="<UNK>", pad_token="<PAD>"):
        """
        Args:
            token_to_idx (dict): a pre-existing map of tokens to indices
            mask_token (str): the MASK token to add into the Vocabulary
            add_unk (bool): whether to add the UNK token
            unk_token (str): the UNK token to add into the Vocabulary
            pad_token (str): the token for padding (to be used in sequences)
        """

        if token_to_idx is None:
            token_to_idx = {}
        self._token_to_idx = token_to_idx

        self._idx_to_token = {idx: token
                              for token, idx in self._token_to_idx.items()}

        self._add_unk = add_unk
        self._unk_token = unk_token
        self._mask_token = mask_token
        self._pad_token = pad_token  # Add pad token

        # Add MASK token and UNK token
        self.mask_index = self.add_token(self._mask_token)
        self.unk_index = -1
        if add_unk:
            self.unk_index = self.add_token(unk_token)

        # Add PAD token
        self.pad_index = self.add_token(self._pad_token)

    def to_serializable(self):
        """ returns a dictionary that can be serialized """
        return {'token_to_idx': self._token_to_idx,
                'add_unk': self._add_unk,
                'unk_token': self._unk_token,
                'mask_token': self._mask_token,
                'pad_token': self._pad_token}

    @classmethod
    def from_serializable(cls, contents):
        """ instantiates the Vocabulary from a serialized dictionary """
        return cls(**contents)

    def add_token(self, token):
        """Update mapping dicts based on the token.

        Args:
            token (str): the item to add into the Vocabulary
        Returns:
            index (int): the integer corresponding to the token
        """
        if token in self._token_to_idx:
            index = self._token_to_idx[token]
        else:
            index = len(self._token_to_idx)
            self._token_to_idx[token] = index
            self._idx_to_token[index] = token
        return index

    def add_many(self, tokens):
        """Add a list of tokens into the Vocabulary

        Args:
            tokens (list): a list of string tokens
        Returns:
            indices (list): a list of indices corresponding to the tokens
        """
        return [self.add_token(token) for token in tokens]

    def lookup_token(self, token):
        """Retrieve the index associated with the token or the UNK index if token isn't present.

        Args:
            token (str): the token to look up
        Returns:
            index (int): the index corresponding to the token
        Notes:
            `unk_index` needs to be >=0 (having been added into the Vocabulary)
              for the UNK functionality
        """
        if self.unk_index >= 0:
            return self._token_to_idx.get(token, self.unk_index)
        else:
            return self._token_to_idx[token]

    def lookup_index(self, index):
        """Return the token associated with the index, handling both single indices and tensors.

        Args:
            index (int or torch.Tensor): the index or tensor of indices to look up
        Returns:
            token (str or list): the token corresponding to the index (or a list of tokens for tensors)
        Raises:
            KeyError: if the index is not in the Vocabulary
        """
        # Check if the input is a tensor (if it's a tensor, we need to handle it accordingly)
        if isinstance(index, torch.Tensor):
            return [self.lookup_index(i.item()) for i in index]  # Use .item() to extract the value from each tensor element

        # Handle single index (non-tensor)
        if index not in self._idx_to_token:
            print(f"Warning: Index {index} not found in vocabulary.")
            return self._token_to_idx.get(self.unk_token, "<UNK>")  # Ensure unk_token is handled properly

        return self._idx_to_token[index]


    def pad_token_index(self):
        """Return the index of the padding token."""
        return self.pad_index  # Return padding token index

    def __str__(self):
        return f"<Vocabulary(size={len(self)})>"

    def __len__(self):
        return len(self._token_to_idx)
```

```python
class SkipGramVectorizer(object):

    def __init__(self, skipgram_vocab):
        """Initialize with the vocabulary"""
        self.skipgram_vocab = skipgram_vocab

    def vectorize(self, target_word, context_words):
        """
        Converts target word and context words to indices.

        Args:
            target_word (str): the target word
            context_words (list of str): list of context words

        Returns:
            tuple: (target_index, context_indices)
        """
        # Convert target word and context words into indices
        target_index = self.skipgram_vocab.lookup_token(target_word)
        context_indices = [self.skipgram_vocab.lookup_token(word) for word in context_words]

        # Ensure context indices is a numpy array and return as numpy array
        return target_index, np.array(context_indices)

    @classmethod
    def from_dataframe(cls, skipgram_df):
        """Build the vocabulary from the entire dataset"""
        skipgram_vocab = Vocabulary(add_unk=True)  # Allow unknown tokens

        for index, row in skipgram_df.iterrows():
            if pd.notna(row.target):
                skipgram_vocab.add_token(str(row.target))  # Convert target to string
            if pd.notna(row.context):
                for token in str(row.context).split(' '):  # Convert context to string
                    skipgram_vocab.add_token(token)

        return cls(skipgram_vocab)


    @classmethod
    def from_serializable(cls, contents):
        """Load a serialized version of the vectorizer"""
        skipgram_vocab = Vocabulary.from_serializable(contents['skipgram_vocab'])
        return cls(skipgram_vocab=skipgram_vocab)

    def to_serializable(self):
        """Serialize the vectorizer"""
        return {'skipgram_vocab': self.skipgram_vocab.to_serializable()}
```

```python
from torch.nn.utils.rnn import pad_sequence

class SkipGramDataset(Dataset):
    def __init__(self, skipgram_df, vectorizer):
        """
        Args:
            skipgram_df (pandas.DataFrame): the dataset
            vectorizer (SkipGramVectorizer): vectorizer instantiated from dataset
        """
        self.skipgram_df = skipgram_df
        self._vectorizer = vectorizer

        # Handle missing context values by replacing NaN with empty strings
        self.skipgram_df['context'] = self.skipgram_df['context'].fillna("")

        # Define the maximum sequence length for the context
        measure_len = lambda context: len(context.split(" "))
        self._max_seq_length = max(map(measure_len, self.skipgram_df.context))

        # Splitting the dataset into train, validation, and test sets
        self.train_df = self.skipgram_df[self.skipgram_df.split == 'train']
        self.train_size = len(self.train_df)

        self.val_df = self.skipgram_df[self.skipgram_df.split == 'val']
        self.validation_size = len(self.val_df)

        self.test_df = self.skipgram_df[self.skipgram_df.split == 'test']
        self.test_size = len(self.test_df)

        # Lookup dictionary to select the dataset split
        self._lookup_dict = {
            'train': (self.train_df, self.train_size),
            'val': (self.val_df, self.validation_size),
            'test': (self.test_df, self.test_size)
        }

        self.set_split('train')

    @classmethod
    def load_dataset_and_make_vectorizer(cls, skipgram_csv):
        """Load dataset and make a new vectorizer from scratch

        Args:
            skipgram_csv (str): location of the dataset
        Returns:
            an instance of SkipGramDataset
        """
        skipgram_df = pd.read_csv(skipgram_csv)

        # Handle missing context values before passing to the vectorizer
        skipgram_df['context'] = skipgram_df['context'].fillna("")

        train_skipgram_df = skipgram_df[skipgram_df.split == 'train']
        return cls(skipgram_df, SkipGramVectorizer.from_dataframe(train_skipgram_df))

    @classmethod
    def load_dataset_and_load_vectorizer(cls, skipgram_csv, vectorizer_filepath):
        """Load dataset and the corresponding vectorizer.
        Used in the case where the vectorizer has been cached for re-use.

        Args:
            skipgram_csv (str): location of the dataset
            vectorizer_filepath (str): location of the saved vectorizer
        Returns:
            an instance of SkipGramDataset
        """
        skipgram_df = pd.read_csv(skipgram_csv)
        vectorizer = cls.load_vectorizer_only(vectorizer_filepath)
        return cls(skipgram_df, vectorizer)

    @staticmethod
    def load_vectorizer_only(vectorizer_filepath):
        """A static method for loading the vectorizer from file.

        Args:
            vectorizer_filepath (str): the location of the serialized vectorizer
        Returns:
            an instance of SkipGramVectorizer
        """
        with open(vectorizer_filepath) as fp:
            return SkipGramVectorizer.from_serializable(json.load(fp))

    def save_vectorizer(self, vectorizer_filepath):
        """Saves the vectorizer to disk using JSON.

        Args:
            vectorizer_filepath (str): the location to save the vectorizer
        """
        with open(vectorizer_filepath, "w") as fp:
            json.dump(self._vectorizer.to_serializable(), fp)

    def get_vectorizer(self):
        """Returns the vectorizer."""
        return self._vectorizer

    def set_split(self, split="train"):
        """Selects the splits in the dataset using a column in the dataframe."""
        self._target_split = split
        self._target_df, self._target_size = self._lookup_dict[split]

    def __len__(self):
        return self._target_size

    def __getitem__(self, index):
        """Fetch the indexed item from the dataset."""
        row = self._target_df.iloc[index]

        target_word = row.target  # Target word
        context_words = str(row.context).split(" ") if pd.notna(row.context) else []  # Ensure it's a list

        # Use the vectorizer to convert the target and context words into indices/vectors
        target_index, context_vector = self._vectorizer.vectorize(target_word, context_words)

        # Convert context_vector (y_target) to tensor
        context_tensor = torch.tensor(context_vector, dtype=torch.long)  # Assuming context_vector is an array

        return {'x_data': target_index,  # Target word index
                'y_target': context_tensor}  # Context word indices as a tensor


    def get_num_batches(self, batch_size):
        """Given a batch size, return the number of batches in the dataset.

        Args:
            batch_size (int)
        Returns:
            number of batches in the dataset
        """
        return len(self) // batch_size


def collate_fn(batch):
    """ Custom collate function to pad variable-length y_target sequences """
    x_data_list = [torch.tensor(sample['x_data'], dtype=torch.long).unsqueeze(0)
                   if isinstance(sample['x_data'], int) else sample['x_data']
                   for sample in batch]

    y_target_list = [torch.tensor(sample['y_target'], dtype=torch.long) for sample in batch]

    # Pad y_target sequences to the max length in the batch
    y_target_padded = pad_sequence(y_target_list, batch_first=True, padding_value=0)

    return {
        'x_data': torch.cat(x_data_list),  # Shape: (batch_size, 1)
        'y_target': y_target_padded  # Shape: (batch_size, max_seq_len_in_batch)
    }

def generate_batches(dataset, batch_size, shuffle=True, drop_last=True, device="cpu"):
    """
    A generator function that uses a custom collate function to handle variable-length sequences.
    """
    dataloader = DataLoader(dataset=dataset, batch_size=batch_size, shuffle=shuffle,
                            drop_last=drop_last, collate_fn=collate_fn)  # Use custom collate_fn

    for batch in dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}  # Move tensors to the correct device
        yield batch
```

```python
class SkipGramClassifier(nn.Module):  # Skip-gram Model
    def __init__(self, vocabulary_size, embedding_size, padding_idx=0):
        """
        Args:
            vocabulary_size (int): number of vocabulary items, controls the
                number of embeddings and prediction vector size
            embedding_size (int): size of the embeddings
            padding_idx (int): default 0; Embedding will not use this index
        """
        super(SkipGramClassifier, self).__init__()

        # Word embedding layer (each word gets an embedding)
        self.embedding = nn.Embedding(num_embeddings=vocabulary_size,
                                      embedding_dim=embedding_size,
                                      padding_idx=padding_idx)

        # Output layer (predicts context words)
        self.fc1 = nn.Linear(in_features=embedding_size,
                             out_features=vocabulary_size)

    def forward(self, x_in, apply_softmax=False):
        """
        Forward pass of the Skip-gram model.
        """
        x_in = x_in.long()  # Convert to LongTensor
        x_embedded = self.embedding(x_in)  # Get word embedding (batch_size, embedding_size)
        x_embedded = F.dropout(x_embedded, 0.3)  # Apply dropout
        y_out = self.fc1(x_embedded)  # Predict multiple context words (batch_size, vocabulary_size)

        if apply_softmax:
            y_out = F.softmax(y_out, dim=1)

        return y_out
```

```python
def make_train_state(args):
    return {'stop_early': False,
            'early_stopping_step': 0,
            'early_stopping_best_val': 1e8,
            'learning_rate': args.learning_rate,
            'epoch_index': 0,
            'train_loss': [],
            'train_acc': [],
            'val_loss': [],
            'val_acc': [],
            'test_loss': -1,
            'test_acc': -1,
            'model_filename': args.model_state_file}

def update_train_state(args, model, train_state):
    """Handle the training state updates.

    Components:
     - Early Stopping: Prevent overfitting.
     - Model Checkpoint: Model is saved if the model is better

    :param args: main arguments
    :param model: model to train
    :param train_state: a dictionary representing the training state values
    :returns:
        a new train_state
    """

    # Save one model at least
    if train_state['epoch_index'] == 0:
        torch.save(model.state_dict(), train_state['model_filename'])
        train_state['stop_early'] = False

    # Save model if performance improved
    elif train_state['epoch_index'] >= 1:
        loss_tm1, loss_t = train_state['val_loss'][-2:]

        # If loss worsened
        if loss_t >= train_state['early_stopping_best_val']:
            # Update step
            train_state['early_stopping_step'] += 1
        # Loss decreased
        else:
            # Save the best model
            if loss_t < train_state['early_stopping_best_val']:
                torch.save(model.state_dict(), train_state['model_filename'])

            # Reset early stopping step
            train_state['early_stopping_step'] = 0

        # Stop early ?
        train_state['stop_early'] = \
            train_state['early_stopping_step'] >= args.early_stopping_criteria

    return train_state

def compute_accuracy(y_pred, y_target):
    _, y_pred_indices = y_pred.max(dim=1)
    n_correct = torch.eq(y_pred_indices, y_target).sum().item()
    return n_correct / len(y_pred_indices) * 100
```

```python
def set_seed_everywhere(seed, cuda):
    np.random.seed(seed)
    torch.manual_seed(seed)
    if cuda:
        torch.cuda.manual_seed_all(seed)

def handle_dirs(dirpath):
    if not os.path.exists(dirpath):
        os.makedirs(dirpath)
```

```python
classifier = classifier.to(args.device)

loss_func = nn.CrossEntropyLoss()
optimizer = optim.Adam(classifier.parameters(), lr=args.learning_rate)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer=optimizer,
                                                 mode='min', factor=0.5,
                                                 patience=1)
train_state = make_train_state(args)

epoch_bar = tqdm(desc='training routine',
                          total=args.num_epochs,
                          position=0)

dataset.set_split('train')
train_bar = tqdm(desc='split=train',
                          total=dataset.get_num_batches(args.batch_size),
                          position=1,
                          leave=True)
dataset.set_split('val')
val_bar = tqdm(desc='split=val',
                        total=dataset.get_num_batches(args.batch_size),
                        position=1,
                        leave=True)

try:
    for epoch_index in range(args.num_epochs):
        train_state['epoch_index'] = epoch_index

        # Iterate over training dataset
        dataset.set_split('train')
        batch_generator = generate_batches(dataset,
                                           batch_size=args.batch_size,
                                           device=args.device)

        running_loss = 0.0
        running_acc = 0.0
        classifier.train()

        for batch_index, batch_dict in enumerate(batch_generator):
            # Step 1: Zero the gradients
            optimizer.zero_grad()

            # Step 2: Compute the output
            y_pred = classifier(x_in=batch_dict['x_data'])

            # Step 3: Compute the loss
            y_target = batch_dict['y_target']

            # Ensure y_target has the correct shape
            if isinstance(loss_func, torch.nn.CrossEntropyLoss):
                y_target = y_target.argmax(dim=1)  # Convert one-hot to class indices
            elif isinstance(loss_func, torch.nn.BCEWithLogitsLoss):
                y_target = y_target.float()  # Convert to float if needed

            loss = loss_func(y_pred, y_target)
            loss_t = loss.item()
            running_loss += (loss_t - running_loss) / (batch_index + 1)

            # Step 4: Use loss to produce gradients
            loss.backward()

            # Step 5: Use optimizer to take gradient step
            optimizer.step()

            # Compute accuracy
            acc_t = compute_accuracy(y_pred, y_target)
            running_acc += (acc_t - running_acc) / (batch_index + 1)

            # Update progress bar
            train_bar.set_postfix(loss=running_loss, acc=running_acc, epoch=epoch_index)
            train_bar.update()

        # Store training loss and accuracy
        train_state['train_loss'].append(running_loss)
        train_state['train_acc'].append(running_acc)

        # Validation phase
        dataset.set_split('val')
        batch_generator = generate_batches(dataset,
                                          batch_size=args.batch_size,
                                          device=args.device)

        running_loss = 0.0
        running_acc = 0.0
        classifier.eval()

        # Disable gradients during validation
        with torch.no_grad():
            for batch_index, batch_dict in enumerate(batch_generator):
                # Step 2: Compute the output
                y_pred = classifier(x_in=batch_dict['x_data'])

                # Step 3: Compute the loss
                y_target = batch_dict['y_target']

                # Ensure y_target has the correct shape for CrossEntropyLoss (1D, class indices)
                if isinstance(loss_func, torch.nn.CrossEntropyLoss):
                    # If y_target is one-hot encoded, convert it to class indices
                    if len(y_target.shape) > 1:  # Check if it's one-hot encoded
                        y_target = y_target.argmax(dim=1)  # Convert to class indices
                elif isinstance(loss_func, torch.nn.BCEWithLogitsLoss):
                    y_target = y_target.float()  # Convert to float if needed for BCE

                # Ensure y_target is now 1D for CrossEntropyLoss
                loss = loss_func(y_pred, y_target)
                loss_t = loss.item()
                running_loss += (loss_t - running_loss) / (batch_index + 1)

                # Compute accuracy
                acc_t = compute_accuracy(y_pred, y_target)
                running_acc += (acc_t - running_acc) / (batch_index + 1)

                # Update progress bar
                val_bar.set_postfix(loss=running_loss, acc=running_acc, epoch=epoch_index)
                val_bar.update()

        # Store validation loss and accuracy
        train_state['val_loss'].append(running_loss)
        train_state['val_acc'].append(running_acc)

        # Update train state and learning rate scheduler
        train_state = update_train_state(args=args, model=classifier,
                                         train_state=train_state)

        # Reduce learning rate if validation loss plateaus
        scheduler.step(train_state['val_loss'][-1])

        # Early stopping
        if train_state['stop_early']:
            break

        train_bar.n = 0
        val_bar.n = 0
        epoch_bar.update()
except KeyboardInterrupt:
    print("Exiting loop")
```

```python
def get_context_words(word, word_to_idx, embeddings, window_size=5, top_n=5):
    """
    Predicts the most likely context words around the target word using Skip-gram.

    Args:
        word (str): The target word.
        word_to_idx (dict): Mapping of words to indices.
        embeddings (torch.Tensor): Word embeddings tensor.
        window_size (int): Number of words to consider before and after the target.
        top_n (int): Number of top context predictions to return.

    Returns:
        list: A list of tuples (score, [context words]).
    """
    if word.lower() not in word_to_idx:
        return f"Word '{word}' is not in the vocabulary."

    word_idx = word_to_idx[word.lower()]
    word_embedding = embeddings[word_idx]

    # Compute similarities to all other words
    similarities = []
    for other_word, idx in word_to_idx.items():
        if other_word == word.lower():
            continue
        other_embedding = embeddings[idx]
        similarity = F.cosine_similarity(word_embedding.unsqueeze(0), other_embedding.unsqueeze(0))
        similarities.append((other_word, similarity.item()))

    # Sort by highest similarity
    sorted_words = sorted(similarities, key=lambda x: x[1], reverse=True)[:top_n * window_size]

    # Group context words into window-sized chunks
    context_predictions = []
    for i in range(0, len(sorted_words), window_size):
        chunk = [w[0] for w in sorted_words[i:i + window_size]]
        score = sum(w[1] for w in sorted_words[i:i + window_size]) / len(chunk)  # Avg similarity score
        context_predictions.append((score, chunk))

    # Sort final output by score
    context_predictions = sorted(context_predictions, key=lambda x: x[0], reverse=True)[:top_n]

    return context_predictions

# Pretty print function
def pretty_print(results):
    """Pretty prints the predicted context words."""
    if isinstance(results, str):
        print(results)
    else:
        for score, words in results:
            print(f"...[{score:.2f}] - {', '.join(words)}")
```
